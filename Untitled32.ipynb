```python
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
# Replace '?' with NaN and convert types if necessary
df.replace('?', pd.NA, inplace=True)
df = df.dropna()  # Or use imputation if preferred
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='target', data=df, palette='Set2')
plt.title('Heart Disease Distribution')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='sex', hue='target', data=df, palette='pastel')
plt.title('Heart Disease by Gender')
plt.xlabel('Sex (1 = Male, 0 = Female)')
plt.ylabel('Count')
plt.legend(title='Heart Disease')
plt.show()
```

```python
sns.histplot(data=df, x='age', hue='target', bins=20, kde=True, palette='coolwarm')
plt.title('Age Distribution by Heart Disease')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='cp', hue='target', data=df, palette='Set3')
plt.title('Chest Pain Type vs Heart Disease')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.legend(title='Heart Disease')
plt.show()
```
